2. **Cream o retea neuronala in care includem doua Hidden Layer(primul strat va avea 3 neuroni, iar cel de-al doilea un singur neuron), iar pentru fiecare neuron din hidden layer consideram ca si functie de activare, functia sigmoid. Vom folosi ca si metoda de ajustare a ponderilor Backpropagation + GD.**

```python
import numpy as np
np.random.seed(1)#ponderile generate random nu se vor schimba la fiecare rulare a programului

def sigmoid(x):#definim functia sigmoid
    return 1 / (1 + np.exp(-x))

def train(X, y, n_hidden, n_epochs, lr):#functia principala care va genera ponderile pentru reteaua neuronala
    n_input = X.shape[1]
    n_output = y.shape[1]

    # initializam ponderile in mod random
    w1 = np.random.randn(n_input, n_hidden)#ponderile pentru layer1-in numar de 6
    w2 = np.random.randn(n_hidden, n_output)#pondderile pentru layer2-in numar de 3
 
    # antrenam modelul pentru n_epochs(n_iteratii)
    cost_values = []#lista ce va contine eroarea pentru fiecare iteratie
    for i in range(n_epochs):
        # Forward pass
        S1 = np.dot(X, w1)#vector ce contine cele trei sume ce intra in hidden layer1
        sigmoid_function = sigmoid(S1)#vector ce retine cele trei valori obtinute dupa aplicarea functiei sigmoid vectorului S1
        S2 = np.dot(sigmoid_function, w2) #este valoarea sumei obtinute in hidden_layer2
        y_pred = sigmoid(S2)#predicted_value

        # compute cost
        cost = np.mean(-y * np.log(y_pred) - (1 - y) * np.log(1 - y_pred))#Cost Function are aceasta forma in cazul distributiei binomiale.
        cost_values.append(cost)#la fiecare iteratiei valoarea pentru cost function este adaugata in lista cost_values

        # backward pass
        error=y_pred - y#calculam eroarea
        w2=w2-np.dot(sigmoid_function.T, error)*lr#update la nivelul ponderilor din hidden_layer2
        d1 = np.dot(error, w2.T) * sigmoid_function * (1 - sigmoid_function)
        w1=w1-np.dot(X.T, d1)*lr#update la nivelul ponderilor din hidden_layer1
        
    return w1, w2, cost_values

# apelam functia de antrenament a modelului pentru valori concrete
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])
w1, w2, cost_values = train(X, y, n_hidden=3, n_epochs=100000, lr=0.1)

#Afisam ponderile in urma "antrenarii modelului"
print("Final weights:")
print(f"w1 = {w1}")
print(f"w2 = {w2}")
print(f"Cost = {cost_values}")
```

```python
x_test = np.array([[1, 1]])
S1_test = np.dot(x_test, w1) 
sigmoid_function_test = sigmoid(S1_test)
S2_test = np.dot(sigmoid_function_test, w2) 
y_pred_test = sigmoid(S2_test)
print(f"Prediction for input {x_test}: {y_pred_test}")
```

3. **Cream o retea neuronala in care includem doua Hidden Layer(primul strat va avea 3 neuroni, iar cel de-al doilea un singur neuron), iar pentru fiecare neuron din hidden layer consideram ca si functie de activare, functia ReLu. Vom folosi ca si metoda de ajustare a ponderilor Backpropagation + GD.**

```python
import numpy as np
np.random.seed(1)#ponderile generate random nu se vor schimba la fiecare rulare a programului

def relu(x):#definim functia Relu
    return (x > 0) * x 

def relu_deriv(x):
    return x>0

def train(X, y, n_hidden, n_epochs, lr):#functia principala care va genera ponderile pentru reteaua neuronala
    n_input = X.shape[1]
    n_output = y.shape[1]

    # initializam ponderile in mod random
    w1 = np.random.randn(n_input, n_hidden)#ponderile pentru layer1-in numar de 6
    w2 = np.random.randn(n_hidden, n_output)#pondderile pentru layer2-in numar de 3
 
    # antrenam modelul pentru n_epochs(n_iteratii)
    cost_values = []#lista ce va contine eroarea pentru fiecare iteratie
    for i in range(n_epochs):
        # Forward pass
        S1 = np.dot(X, w1)#vector ce contine cele trei sume ce intra in hidden layer1
        relu_function = relu(S1)#vector ce retine cele trei valori obtinute dupa aplicarea functiei ReLu vectorului S1
        S2 = np.dot(relu_function, w2) #este valoarea sumei obtinute in hidden_layer2
        y_pred = relu(S2)#predicted_value

        # compute cost
        cost = np.mean((y_pred-y)**2)#Cost Function are aceasta forma in cazul distributiei normale.
        cost_values.append(cost)#la fiecare iteratiei valoarea pentru cost function este adaugata in lista cost_values

        # backward pass
        error=y_pred - y#calculam eroarea
        w2=w2-np.dot(relu_function.T, error)*lr#update la nivelul ponderilor din hidden_layer2
        d1 = np.dot(error, w2.T) * relu_deriv(S1)
        w1=w1-np.dot(X.T, d1)*lr#update la nivelul ponderilor din hidden_layer1
        
    return w1, w2, cost_values

# apelam functia de antrenament a modelului pentru valori concrete
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])
w1, w2, cost_values = train(X, y, n_hidden=3, n_epochs=100, lr=0.1)

#Afisam ponderile in urma "antrenarii modelului"
print("Final weights:")
print(f"w1 = {w1}")
print(f"w2 = {w2}")
print(f"Cost = {cost_values}")
```

```python
x_test = np.array([[0, 1]])
S1_test = np.dot(x_test, w1) 
relu_function_test = relu(S1_test)
S2_test = np.dot(relu_function_test, w2) 
y_pred_test = relu(S2_test)
print(f"Prediction for input {x_test}: {y_pred_test}")
```
